```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import pandas as pd

# Descargar los recursos necesarios de NLTK (solo necesario si no están descargados previamente)
nltk.download('punkt')
nltk.download('stopwords')

# DataFrame
from google.colab import drive
drive.mount('/content/drive')
movies_data = pd.read_csv("/content/drive/MyDrive/dataIA/movies/finalMovies5K.csv")
```

```python
def extract_keywords(title, overview, keywords):
    if keywords != '[]':  # Verificar si la lista de "keywords" está vacía
        # Tokenizar el resumen de la película
        tokens = word_tokenize(overview.lower())  # Convertir el texto a minúsculas y tokenizar

        # Filtrar palabras vacías (stopwords)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

        # Calcular la frecuencia de las palabras
        fdist = FreqDist(filtered_tokens)

        # Obtener las palabras más frecuentes (keywords)
        most_common_words = fdist.most_common(3)  # Obtener las 3 palabras más frecuentes

        # Extraer las palabras clave de la lista de tuplas (palabra, frecuencia)
        most_common_keywords = [word for word, freq in most_common_words]
        print(f"Film {title} keywords generated: {most_common_keywords}")
        return most_common_keywords
    else:
        keywords_list = eval(keywords)
        return keywords_list
movies_data['keywords'] = movies_data.apply(lambda row: extract_keywords(row['title'], row['overview'], row['keywords']), axis=1)
```
