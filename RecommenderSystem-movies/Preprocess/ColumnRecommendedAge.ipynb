```python
from dotenv import load_dotenv
import os
import openai
import pandas as pd

# Cargar variables de entorno
load_dotenv()
```

```python
# Establece tu API Key de OpenAI
openai.api_key = os.getenv("OPENAI_API_KEY")

# Cargar el dataset movies_with_genres
movies_with_genres = pd.read_csv('movies_with_genres.csv', error_bad_lines=False)

# Función para obtener la edad recomendada utilizando GPT
def get_recommended_age(release_date):
    # Formular la consulta para GPT
    query = f"¿Para qué edad está recomendada una película estrenada en {release_date}?"

    # Hacer la consulta a GPT
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=query,
        max_tokens=50,
        temperature=0.5
    )

    # Extraer la respuesta de GPT
    recommended_age = response.choices[0].text.strip()
    return recommended_age
```

```python
# Aplicar la función a la columna "release_date" del dataset
movies_with_genres['recommendedAge'] = movies_with_genres['release_date'].apply(get_recommended_age)
```

```python
# Guardar el dataset actualizado
movies_with_genres.to_csv('movies_with_genres_updated.csv', index=False)
```
