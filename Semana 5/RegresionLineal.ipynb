```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Generar datos
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

X_nuevos = np.linspace(0, 2, 1000).reshape(-1, 1)
y_nuevos = modelo.predict(X_nuevos)
```

```python
nuevo_X = np.array([[1830]])  # Valor de X para predecir y
y_pred = modelo.predict(nuevo_X)

print("El valor de y predicho para X = 1830 es:", y_pred[0][0])
```

```python
import matplotlib.pyplot as plt

# Graficar los datos y la línea de regresión
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X_nuevos, y_nuevos, color='red', label='Línea de Regresión')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```
